```python
import seaborn as sns
import pandas as pd
# The first column shows the plots for the number of fatal police shootings where the victim was armed per county
# against the number of fatal police shootings that were X race per county
df = pd.read_csv("countiesPerArmed.csv")

sns.pairplot(df, corner=True)
```

```python
# Here the same thing is done again with the number of fatal police shootings where the victim was armed that is either
# against the number of fatal police shootings that were X race per county
# Black (first column) or Native American (second column) 
df = pd.read_csv("countiesPerArmedRaceSelected.csv")

sns.pairplot(df,kind="scatter",corner=True)
```

```python
# Hispanic (first column), White (second column)
df = pd.read_csv("countiesPerArmedRaceSelected2.csv")

sns.pairplot(df,kind="scatter",corner=True)
```
